**EDA(Exploratory Data Analysis)**

```python
sales_platform = df.groupby("Platform")["TotalAmount"].sum().sort_values(ascending=False)

plot_no = 1
plt.figure(figsize=(7,6))
sales_platform.plot(kind="bar",colormap = 'Set2',edgecolor ="black",)
plt.title(f'{plot_no}.Revenue of platform' , color = 'Red')
plt.xlabel("Platform",color='blue')
plt.ylabel("Revenue",color='blue')
plt.legend()
plt.show()
```

```python
category_revenue =  df.groupby("Category")["TotalAmount"].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
category_revenue.plot(kind="bar",colormap = 'viridis',width = 0.7,edgecolor ="black")
plt.title(f'{plot_no}.Revenue of Platform' , color = 'Red')
plt.xlabel("Category",color = 'blue')
plt.ylabel("Total Amoubnt",color = 'blue')
plt.legend()
plt.show()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='City', color= 'blue')
plt.title(f'{plot_no}. Order Volume by City')
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.heatmap(df[['Price','Quantity','TotalAmount','Rating','Reviews']].corr(), annot=True)
plt.title(f'{plot_no}. Correlation Heatmap of Numerical Features')

plot_no += 1
```

```python
category_revenue = df.groupby("Category")["TotalAmount"].sum()
plt.figure(figsize=(7,7))
plt.pie(
  category_revenue,
    labels=category_revenue.index,
    autopct='%1.1f%%',
    startangle=90,
  
)
plt.title("Revenue share by platform")
plt.show()
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Platform', hue='Category')
plt.title(f'{plot_no}. Category Mix Across Platforms')

plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['Price'], kde=True)   # Kde :- (Kernel Density Estimation)
plt.title(f'{plot_no}. Price Distribution of All Products')

plot_no += 1
```
